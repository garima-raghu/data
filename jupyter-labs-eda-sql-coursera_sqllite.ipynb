### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
# Execute the SQL query to retrieve the unique launch sites
cur.execute("SELECT DISTINCT Launch_Site FROM SPACEXTBL")

results = cur.fetchall()

# Iterate over the rows and print the data
for row in results:
    print(row)
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
cur.execute("SELECT * FROM SPACEXTBL WHERE Launch_Site LIKE 'CCA%' LIMIT 5")

# Fetch all rows from the result set
results = cur.fetchall()

# Iterate over the rows and print the data
for row in results:
    print(row)
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
# Execute the PRAGMA statement to get column information
cur.execute("PRAGMA table_info(SPACEXTBL)")

# Fetch all the rows returned by the query
rows = cur.fetchall()

# Extract the column names from the rows
column_names = [row[1] for row in rows]

# Print the column names
print("Column names of table SPACEXTBL:")
for column_name in column_names:
    print(column_name)

```

```python
# Execute the SELECT query
cur.execute("SELECT SUM(PAYLOAD_MASS__KG_) AS TotalPayloadMass FROM SPACEXTBL WHERE Customer = 'NASA (CRS)'")

# Fetch the result
result = cur.fetchone()

# Extract the total payload mass
total_payload_mass = result[0]

# Print the total payload mass
print("Total Payload Mass carried by NASA (CRS):", total_payload_mass)
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
# Execute the SELECT query
cur.execute("SELECT AVG(PAYLOAD_MASS__KG_) AS AveragePayloadMass FROM SPACEXTBL WHERE Booster_Version = 'F9 v1.1'")

# Fetch the result
result = cur.fetchone()

# Extract the average payload mass
average_payload_mass = result[0]

# Print the average payload mass
print("Average Payload Mass carried by booster version F9 v1.1:", average_payload_mass)
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
# Execute the SQL query
cur.execute("SELECT MIN(Date) FROM SPACEXTBL WHERE \"Landing_Outcome\" LIKE 'Success (ground pad)%'")

# Fetch the result
result = cur.fetchone()

# Print the date
print("Date of first successful landing outcome in ground pad:", result[0])
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
# Execute the SQL query
cur.execute("SELECT DISTINCT Booster_Version FROM SPACEXTBL WHERE \"Landing_Outcome\" = 'Success' AND PAYLOAD_MASS__KG_ > 4000 AND PAYLOAD_MASS__KG_ < 6000")

# Fetch the results
results = cur.fetchall()

# Print the booster names
print("Boosters with successful landing on drone ship and payload mass between 4000 and 6000:")
for row in results:
    print(row[0])
```

### Task 7

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 4, 2) as month to get the months and substr(Date,7,4)='2015' for year.**


```python
# Execute the SQLite query
cur.execute("""
    SELECT
        CASE substr(Date, 4, 2)
            WHEN '01' THEN 'January'
            WHEN '02' THEN 'February'
            WHEN '03' THEN 'March'
            WHEN '04' THEN 'April'
            WHEN '05' THEN 'May'
            WHEN '06' THEN 'June'
            WHEN '07' THEN 'July'
            WHEN '08' THEN 'August'
            WHEN '09' THEN 'September'
            WHEN '10' THEN 'October'
            WHEN '11' THEN 'November'
            WHEN '12' THEN 'December'
            ELSE 'Unknown'
        END AS Month,
        landing_outcome,
        booster_version,
        launch_site
    FROM
        SPACEXTBL
    WHERE
        substr(Date, 7, 4) = '2015'
        AND landing_outcome = 'Failure (drone ship)'
""")

# Fetch all the rows returned by the query
rows = cur.fetchall()

# Print the column names
print("Month\t\tLanding Outcome\t\tBooster Version\t\tLaunch Site")
for row in rows:
    month = row[0]
    landing_outcome = row[1]
    booster_version = row[2]
    launch_site = row[3]
    print(f"{month}\t\t{landing_outcome}\t\t{booster_version}\t\t{launch_site}")
```

### Task 10




##### Rank the  count of  successful landing_outcomes between the date 04-06-2010 and 20-03-2017 in descending order.


```python
# Execute the SQLite query
cur.execute("""
    SELECT
        landing_outcome,
        COUNT(*) AS successful_landings
    FROM
        SPACEXTBL
    WHERE
        Date BETWEEN '2010-06-04' AND '2017-03-20'
        AND landing_outcome = 'Success (drone ship)'
    GROUP BY
        landing_outcome
    ORDER BY
        successful_landings DESC
""")

# Fetch all the rows returned by the query
rows = cur.fetchall()

# Print the result
print("Landing Outcome\t\tSuccessful Landings")
for row in rows:
    landing_outcome = row[0]
    successful_landings = row[1]
    print(f"{landing_outcome}\t\t\t{successful_landings}")

```
